```python
import numpy as np
import matplotlib.pyplot as plt 
x0 = np.pi/6
band = np.arange(start = 1, stop = 22)
steps = 2 / (2 ** band)
```

```python
# methods
def method1(foo, h):
    return (foo(x0 + h) - foo(x0)) / h

def method2(foo, h):
    return (foo(x0) - foo(x0 - h)) / h

def method3(foo, h):
    return (foo(x0 + h) - foo(x0 - h)) / (2 * h)

def method4(foo, h):
    return ((4 / 3) * method3(foo, h)) - ((1 / 3) * (foo(x0 + 2 * h) - foo(x0 - 2 * h)) / (4 * h))

def method5(foo, h):
    return ((3 / 2) * method3(foo, h)) - ((3 / 5) * (foo(x0 + 2 * h) - foo(x0 - 2 * h)) / (4 * h)) + ((1 / 10) * (foo(x0 + 3 * h) - foo(x0 - 3 * h)) / (6 * h))
```

```python
# functions

def f1(x):
    return np.sin(x ** 2)

def f2(x):
    return np.cos(np.sin(x))

def f3(x):
    return np.exp(np.sin(np.cos(x)))

def f4(x):
    return np.log(x + 3)

def f5(x):
    return ((x + 3) ** 0.5)
```

```python
# derivatives of functions

der1 = 2 * x0 * np.cos(x0 ** 2)

der2 = -np.sin(np.sin(x0)) * np.cos(x0)

der3 = f3(x0) * np.cos(np.cos(x0)) * (-np.sin(x0))

der4 = 1 / (x0 + 3)

der5 = (1 / 2) / (np.sqrt(x0 + 3))
```

```python
def draw(foo, der, title):
    y1 = np.abs(der - method1(foo, steps))
    y2 = np.abs(der - method2(foo, steps))
    y3 = np.abs(der - method3(foo, steps))
    y4 = np.abs(der - method4(foo, steps))
    y5 = np.abs(der - method5(foo, steps))

    plt.figure(figsize=[12, 5], dpi = 100)
    plt.plot(steps, y1, label='method 1')
    plt.plot(steps, y2, label='method 2')
    plt.plot(steps, y3, label='method 3')
    plt.plot(steps, y4, label='method 4')
    plt.plot(steps, y5, label='method 5')    
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel("$log_h$")
    plt.ylabel("$log y$")
    plt.legend(loc='best', bbox_to_anchor=(1, 1))
    plt.title(title)
    plt.grid()
    plt.show() 
```

```python
draw(f1, der1, "Errors for the first function")
```

```python
draw(f2, der2, "Errors for the second function")
```

```python
draw(f3, der3, "Errors for the third function")
```

```python
draw(f4, der4, "Errors for the fourth function")
```

```python
draw(f5, der5, "Errors for the fifth function")
```
